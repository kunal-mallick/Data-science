### **HISTOGRAM**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### **MATPLOTLIB** **LIBRARY**

```python
df = pd.read_csv("/content/train.csv")
plt.rcParams['figure.figsize'] = (5,3)
plt.rcParams['figure.dpi'] = 300
df
```

```python
m = df["Age"].mean()
md = df["Age"].median()

plt.hist(df["Age"], edgecolor = 'black')

plt.xlabel("mean > median = +Ve skewness \n median > mean = -Ve skewness \n mean = median = normal distribution ")
plt.ylabel('frequency')
plt.axvline(m, color= 'green', label = 'mean')
plt.axvline(md, color= 'purple', label = 'median')
plt.grid()
plt.legend()
plt.show()
```
